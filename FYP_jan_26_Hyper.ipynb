```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# Fetch dataset
bank_marketing = fetch_ucirepo(id=222)

# Data as pandas dataframes
X = bank_marketing.data.features
y = bank_marketing.data.targets

# Combine for easier inspection and EDA
df = pd.concat([X, y], axis=1)
```

# **Inspect the Dataset**
This step helps you understand the structure, data types, and identify missing values.

```python
# View first few rows
print(df.head())

# Check data types and non-null counts
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

# **Preprocessing**
The Bank Marketing dataset often uses "unknown" as a placeholder for missing values. We also need to encode categorical variables for modeling.

```python
import pandas as pd
import numpy as np

# 1. Handle "unknown" values by converting to NaNs for standard processing
df.replace('unknown', np.nan, inplace=True)

# 2. Binary Encoding for the target variable 'y'
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# 3. Targeted Missing Value Imputation
# For columns with very high null counts (poutcome), it's safer to categorize them
df['poutcome'] = df['poutcome'].fillna('not_contacted')

# For others, we use the Mode (most frequent value)
cols_to_fix = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact']
for col in cols_to_fix:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

# 4. Drop redundant or constant columns
# We drop 'was_contacted' because it has zero variance (constant value)
# We often drop 'duration' for predictive modeling because it's not known before a call
df.drop(columns=['was_contacted'], inplace=True, errors='ignore')

# 5. Final Verification
print("Missing values after preprocessing:")
print(df.isnull().sum())
print(f"\nRemaining columns: {df.columns.tolist()}")
```

# **Data Preprocessing for Modeling**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# 1. Separate Features and Target
X = df.drop('y', axis=1)
y = df['y']

# 2. Identify column types
# Numerical columns need scaling; Categorical columns need encoding
num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object']).columns.tolist()

# 3. Create a Transformer for the pipeline
# This applies OneHotEncoding to cats and Scaling to nums simultaneously
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)
    ])

# 4. Apply the transformations
X_processed = preprocessor.fit_transform(X)

# 5. Get feature names (optional, for readability)
cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(cat_cols)
all_feature_names = num_cols + list(cat_feature_names)

# Convert back to DataFrame for a clean view
X_final = pd.DataFrame(X_processed, columns=all_feature_names)

print("Preprocessed Data Head:")
print(X_final.head())
```

# **Model Training**

## **Model Initialization**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

# Split the processed data
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}
```

## **Model Training and Prediction**

```python
import time

results = {}

for name, model in models.items():
    # Training Time
    start_train = time.time()
    model.fit(X_train, y_train)
    end_train = time.time()

    # Prediction/Testing Time
    start_test = time.time()
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    end_test = time.time()

    # Store predictions and times for the next cell
    results[name] = {
        'y_pred': y_pred,
        'y_prob': y_prob,
        'train_time': end_train - start_train,
        'test_time': end_test - start_test
    }
    print(f"Finished training {name}")
```

## **Initial Model Evaluation**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

evaluation_metrics = []

for name, data in results.items():
    y_pred = data['y_pred']
    metrics = {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "Train Time (s)": data['train_time'],
        "Test Time (s)": data['test_time']
    }
    evaluation_metrics.append(metrics)

# Display as a table
metrics_df = pd.DataFrame(evaluation_metrics)
print(metrics_df)
```

## **Model Visualizations**

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

for i, (name, data) in enumerate(results.items()):
    # Confusion Matrix
    cm = confusion_matrix(y_test, data['y_pred'])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=axes[0, i], cmap='Blues', colorbar=False)
    axes[0, i].set_title(f"Confusion Matrix: {name}")

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, data['y_prob'])
    roc_auc = auc(fpr, tpr)
    axes[1, i].plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
    axes[1, i].plot([0, 1], [0, 1], 'k--')
    axes[1, i].set_title(f"ROC Curve: {name}")
    axes[1, i].legend(loc='lower right')

plt.tight_layout()
plt.show()
```

# **Hyperparameter Optimization**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Faster Parameter Grids
param_grids = {
    "Logistic Regression": {
        'C': [0.1, 1, 10],
        'solver': ['lbfgs']
    },
    "Fast SVM": {
        'estimator__C': [0.1, 1, 10] # Corrected parameter name
    },
    "XGBoost": {
        'n_estimators': [100, 150],
        'learning_rate': [0.1, 0.2],
        'max_depth': [3, 5],
        'subsample': [0.8]
    }
}

# 2. Models Initialization
# CalibratedClassifierCV is used with LinearSVC to enable predict_proba for ROC curves
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Fast SVM": CalibratedClassifierCV(LinearSVC(dual=False, random_state=42)),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

final_results = {}

for name, model in models.items():
    print(f"--- Optimizing {name} ---")

    # Using a slightly lower n_iter and cv for speed
    rs = RandomizedSearchCV(
        model,
        param_distributions=param_grids[name],
        n_iter=3,
        cv=3,
        scoring='f1',
        n_jobs=-1,
        random_state=42
    )

    # Training Time
    start_train = time.time()
    rs.fit(X_train, y_train)
    best_model = rs.best_estimator_
    end_train = time.time()

    # Testing Time
    start_test = time.time()
    y_pred = best_model.predict(X_test)
    y_prob = best_model.predict_proba(X_test)[:, 1]
    end_test = time.time()

    # Store Results
    final_results[name] = {
        'model': best_model,
        'y_pred': y_pred,
        'y_prob': y_prob,
        'train_time': end_train - start_train,
        'test_time': end_test - start_test,
        'best_params': rs.best_params_
    }
    print(f"Best Parameters for {name}: {rs.best_params_}\n")

# 3. Comprehensive Evaluation Table
eval_data = []
for name, res in final_results.items():
    eval_data.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, res['y_pred']),
        "Precision": precision_score(y_test, res['y_pred']),
        "Recall": recall_score(y_test, res['y_pred']),
        "F1-Score": f1_score(y_test, res['y_pred']),
        "Train Time (s)": f"{res['train_time']:.4f}",
        "Test Time (s)": f"{res['test_time']:.4f}"
    })

print("--- Final Performance Metrics ---")
metrics_df = pd.DataFrame(eval_data)
print(metrics_df.to_string(index=False))

# 4. Visualization: Confusion Matrix & ROC Curve
fig, axes = plt.subplots(2, 3, figsize=(20, 12))

for i, (name, res) in enumerate(final_results.items()):
    # Confusion Matrix
    cm = confusion_matrix(y_test, res['y_pred'])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=axes[0, i], cmap='viridis', colorbar=False)
    axes[0, i].set_title(f"CM: {name}")

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, res['y_prob'])
    roc_auc = auc(fpr, tpr)
    axes[1, i].plot(fpr, tpr, color='blue', lw=2, label=f'AUC = {roc_auc:.2f}')
    axes[1, i].plot([0, 1], [0, 1], color='red', linestyle='--')
    axes[1, i].set_title(f"ROC: {name}")
    axes[1, i].legend(loc="lower right")

plt.tight_layout()
plt.show()
```

## **Cross-Validation**

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
import pandas as pd
import numpy as np

# 1. Initialize models with YOUR specific best parameters
optimized_models = {
    "Logistic Regression": LogisticRegression(
        solver='lbfgs',
        C=10,
        max_iter=1000,
        random_state=42
    ),
    "Fast SVM": CalibratedClassifierCV(
        LinearSVC(C=0.1, dual=False, random_state=42)
    ),
    "XGBoost": XGBClassifier(
        subsample=0.8,
        n_estimators=150,
        max_depth=3,
        learning_rate=0.2,
        eval_metric='logloss',
        random_state=42
    )
}

# 2. Setup K-Fold (5 folds is standard for a balance of speed and accuracy)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ['accuracy', 'f1', 'precision', 'recall']

cv_results_final = []

print("Running K-Fold Cross Validation...")

# Ensure the target variable is explicitly numerical before starting CV
# This handles potential kernel state inconsistencies where df['y'] might revert
y_target_cv = df['y'].map({'yes': 1, 'no': 0}) if df['y'].dtype == 'object' else df['y']

for name, model in optimized_models.items():
    print(f"Processing {name}...")

    # Perform Cross Validation, using the explicitly prepared numerical y_target_cv
    cv_output = cross_validate(model, X_processed, y_target_cv, cv=skf, scoring=scoring, n_jobs=-1)

    # Calculate Mean and Standard Deviation for each metric
    cv_results_final.append({
        "Model": name,
        "Accuracy": f"{cv_output['test_accuracy'].mean():.4f} (±{cv_output['test_accuracy'].std():.4f})",
        "F1-Score": f"{cv_output['test_f1'].mean():.4f} (±{cv_output['test_f1'].std():.4f})",
        "Precision": f"{cv_output['test_precision'].mean():.4f} (±{cv_output['test_precision'].std():.4f})",
        "Recall": f"{cv_output['test_recall'].mean():.4f} (±{cv_output['test_recall'].std():.4f})"
    })

# 3. Display the Results
cv_df = pd.DataFrame(cv_results_final)
print("\n--- 5-Fold Cross Validation Results (Mean ± Std Dev) ---")
print(cv_df.to_string(index=False))
```
